```
from catboost import CatBoostRegressor
import lightgbm as lgb
import xgboost as xgb
from sklearn import model_selection
import sklearn
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```
data=pd.read_csv("clean_movie.csv")
```

```
data.head(10)
```

```


X = data.drop(columns=['revenue'])
X = X.drop(columns=['Unnamed: 0'])
Y = data['revenue']
```

```
X.head()
```

```
Y.head()
```

```
validation_size = 0.2
seed = 100
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

```
model_XG = xgb.XGBRegressor() 
model_cb=CatBoostRegressor()
model_lgb=lgb.LGBMRegressor()
model_XG.fit(X_train.values, Y_train.values)
model_cb.fit(X_train.values, Y_train.values)
model_lgb.fit(X_train.values, Y_train.values)
```

```
def predict_test(X_train,Y_train):
  trainResult_XG = model_XG.predict(X_train.values)
  trainResult_lg = model_lgb.predict(X_train.values)
  trainResult_cb = model_cb.predict(X_train.values)
  
  print("Revenue score error from XGBOOST Regressor")
  print("Mean squared error: %.5f"%np.sqrt( mean_squared_error(Y_train.values, trainResult_XG)))
  print('Variance score: %.4f' % r2_score(Y_train.values, trainResult_XG))
  print("\n")
  print("Revenue score error from LGBM Regressor")
  print("Mean squared error: %.5f"%np.sqrt( mean_squared_error(Y_train.values, trainResult_lg)))
  print('Variance score: %.4f' % r2_score(Y_train.values, trainResult_lg))
  print("\n")
  print("Revenue score error from CATBoost Regressor")
  print("Mean squared error: %.5f"%np.sqrt( mean_squared_error(Y_train.values, trainResult_cb)))
  print('Variance score: %.4f' % r2_score(Y_train.values, trainResult_cb))
  print("=========================================================================")
  return trainResult_XG,trainResult_lg,trainResult_cb
  
```

```
predict_test(X_train,Y_train)
predict_test(X_test,Y_test)
```

```
def predict(X_train):
  trainResult_XG = model_XG.predict(X_train)
  trainResult_lg = model_lgb.predict(X_train)
  trainResult_cb = model_cb.predict(X_train)
  return trainResult_XG,trainResult_lg,trainResult_cb
```
